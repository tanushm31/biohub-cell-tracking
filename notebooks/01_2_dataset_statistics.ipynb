```python
from src.data.dataloader import BioHubDataset
from src.analysis.dataset_stats import analyze_cell_counts
from matplotlib import pyplot as plt
dataset = BioHubDataset("../data/raw/train")

stats = analyze_cell_counts(dataset)
```

```python
print(f"Average : {stats.average:.2f}")
print(f"Minimum : {stats.minimum}")
print(f"Maximum : {stats.maximum}")
```

```python
plt.figure(figsize=(8,4))

plt.hist(
    stats.cells_per_frame,
    bins=25,
)

plt.title("Annotated Cells per Frame")
plt.xlabel("Cells")
plt.ylabel("Frequency")
```
